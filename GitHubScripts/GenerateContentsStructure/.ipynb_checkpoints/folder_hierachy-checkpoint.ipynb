```python
# folders only RECOMMENDED

import os
 
root_path = "C:\\Users\\\FPSScripting2\\Documents\\Consolidated_Models"
level = "\t"
 
def get_subdir_list(path):
    return [element for element in os.listdir(path) if os.path.isdir(os.path.join(path, element))]
 
 
def print_children(path, level):
    my_children = get_subdir_list(path)
    for child in my_children:
        print(level + child)
        print_children(os.path.join(path, child), level+"\t")
 
print_children(root_path, level)
```

```python
# include files NOT RECOMMENDED

import os
 
root_path = "C:\\Users\\\FPSScripting2\\Documents\\Consolidated_Models"
level = "\t"
 
def get_subdir_list(path)
    return os.listdir(path)
 
 
def print_children(path, level):
    my_children = get_subdir_list(path)
    for child in my_children:
        print(level + child)
        full_path_of_child = os.path.join(path, child)
        if os.path.isdir(full_path_of_child):
            print_children(full_path_of_child, level+"\t")
 
print_children(root_path, level)
```
